## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.

- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

```python
df_data1['dteday'] = pd.to_datetime(df_data1['dteday'])
df_data1.info()
```

```python
df_data2['dteday'] = pd.to_datetime(df_data2['dteday'])
df_data2.info()
```

```python
new_column_names = {
    'dteday': 'date',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_condition',
    'temp': 'temperature',
    'atemp': 'atemp',
    'hum': 'humidity',
    'windspeed': 'wind_speed',
    'casual': 'casual_users',
    'registered': 'registered_users',
    'cnt': 'total_rentals'
}

df_data1.rename(columns=new_column_names, inplace=True)
df_data2.rename(columns={**new_column_names, 'hr': 'hour'}, inplace=True)

# Mapping for categories
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
year_map = {0: '2011', 1: '2012'}
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
weekday_map = {0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'}
weather_map = {1: 'Clear', 2: 'Misty/Cloudy', 3: 'Light Snow/Rain', 4: 'Heavy Rain/Ice'}

# Apply mapping to df_data1
df_data1['season'] = df_data1['season'].map(season_map)
df_data1['year'] = df_data1['year'].map(year_map)
df_data1['month'] = df_data1['month'].map(month_map)
df_data1['weekday'] = df_data1['weekday'].map(weekday_map)
df_data1['weather_condition'] = df_data1['weather_condition'].map(weather_map)

# Apply mapping to df_data2
df_data2['season'] = df_data2['season'].map(season_map)
df_data2['year'] = df_data2['year'].map(year_map)
df_data2['month'] = df_data2['month'].map(month_map)
df_data2['weekday'] = df_data2['weekday'].map(weekday_map)
df_data2['weather_condition'] = df_data2['weather_condition'].map(weather_map)

print("Data cleaning selesai. Contoh data df_data1:")
display(df_data1.head())
```

**Insight:**
- **Pembaruan Nama Kolom:** Seluruh nama kolom yang disingkat (seperti `yr`, `mnth`, `cnt`) telah diubah menjadi format yang lebih deskriptif (`year`, `month`, `total_rentals`) untuk meningkatkan keterbacaan kode dan analisis.
- **Transformasi Tipe Data:** Kolom `date` telah berhasil dikonversi ke tipe data `datetime`, memungkinkan analisis berbasis deret waktu (time-series).
- **Pemetaan Kategorikal:** Nilai numerik pada kolom `season`, `year`, `month`, `weekday`, dan `weather_condition` telah dipetakan ke label string yang relevan. Hal ini mempermudah interpretasi data saat melakukan visualisasi nanti.

## Analisis Lanjutan (Opsional)
Pada bagian ini, dilakukan analisis RFM untuk melihat ringkasan aktivitas penyewaan serta *Manual Clustering* dengan teknik *binning* pada variabel suhu untuk melihat segmentasi penggunaan sepeda berdasarkan kondisi termal.

```python
# --- 1. RFM Analysis (Sederhana) ---
# Karena data ini bersifat agregat harian, kita akan mengasumsikan:
# Recency: Selisih hari terakhir data dengan hari terakhir transaksi pengguna
# Frequency: Berapa kali ada transaksi (jumlah baris/hari)
# Monetary: Total rentals (total_rentals)

last_date = df_data1['date'].max()

# RFM untuk dataset keseluruhan
rfm_df = df_data1.agg({
    'date': lambda x: (last_date - x.max()).days,
    'instant': 'count',
    'total_rentals': 'sum'
}).reset_index()

rfm_df.columns = ['metric', 'value']
print("Ringkasan RFM (Dataset Keseluruhan):")
display(rfm_df)

```

```python
# --- 2. Manual Clustering (Binning) ---
# Mengelompokkan suhu menjadi kategori: Cold, Mild, Warm, Hot
bins = [0, 0.3, 0.5, 0.7, 1.0]
labels = ['Cold', 'Mild', 'Warm', 'Hot']
df_data1['temp_category'] = pd.cut(df_data1['temperature'], bins=bins, labels=labels)

temp_cluster = df_data1.groupby('temp_category', observed=True)['total_rentals'].mean().reset_index()
print("\nRata-rata Penyewaan berdasarkan Klaster Suhu (Manual Binning):")
display(temp_cluster)
```

**Insight:**

1. **RFM Analysis (Ringkasan):**
   - **Recency:** Nilai 0 menunjukkan bahwa transaksi terakhir terjadi tepat di hari terakhir dataset diambil, menandakan sistem penyewaan masih aktif dan konsisten hingga akhir periode.
   - **Frequency & Monetary:** Total penyewaan mencapai lebih dari 3,2 juta dalam 731 hari, menunjukkan volume transaksi yang sangat besar dengan rata-rata ribuan penyewaan per hari.

2. **Manual Clustering (Suhu):**
   - Terdapat tren kenaikan penyewaan yang linear terhadap kategori suhu.
   - Klaster **'Hot'** dan **'Warm'** memiliki rata-rata penyewaan tertinggi (keduanya di atas 5.500), yang mengonfirmasi bahwa cuaca hangat adalah pendorong utama aktivitas penyewaan.
   - Klaster **'Cold'** memiliki rata-rata terendah (~2.381), menunjukkan penurunan permintaan hingga lebih dari 50% dibandingkan saat cuaca panas, yang memerlukan strategi khusus seperti promo musim dingin.
